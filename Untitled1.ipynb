## 註解

```python
a=3 #單行註解
b=5
print(a+b)
```

```python
a=5
b=3
#透過第三變數c, 交換a, b的值
c=a
a=b
b=c
print(a)
print(b)
```

```python
#不透過第三變數, 交換a,b的值
a=5
b=3
a=a+b
b=a-b
a=a-b
print(a)
print(b)
```

```python
a=5
b=3
a,b=b,a
print(a)
print(b)
```

# 變數
## 變數的平行設定

```python
x, y, z = 3, 4, 5
print(x,y,z)
```

```python
x, y, z = 3, 4, 5
x +=1 #x=x+1
y *=2 #y=y*2
z **=3 #z=z^3
print(x, y, z)
```
